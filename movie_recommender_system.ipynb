```python
import ast
```

```python
def string_of_list_to_list(obj):
    List=[]
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List
```

```python
movies['genres']=movies['genres'].apply(string_of_list_to_list)
```

```python
def cast_selection(obj):
    List=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter<5:
            List.append(i['name'])
        counter+=1
    return List
   
```

```python
movies['cast']=movies['cast'].apply(cast_selection)
```

```python
def movie_director(obj):
    List=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            List.append(i['name'])
    return List
```

```python
movies['crew'] = movies['crew'].apply(movie_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
def space_removal(obj):
    L=[]
    for i in obj:
        L.append(i.replace(" ",""))
    return L
```

```python
movies['genres'] = movies['genres'].apply(space_removal)
movies['keywords'] = movies['keywords'].apply(space_removal)
movies['cast'] = movies['cast'].apply(space_removal)
movies['crew'] = movies['crew'].apply(space_removal)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
# Download necessary resources
nltk.download('punkt')
nltk.download('wordnet')
```

```python
# Function to lemmatize text
def lemmatize_text(text):
    # Tokenize the text
    tokens = word_tokenize(text)
    # Initialize lemmatizer
    lemmatizer = WordNetLemmatizer()
    # Lemmatize each token
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    # Join tokens back into a single string
    lemmatized_text = ' '.join(lemmatized_tokens)
    return lemmatized_text
```

```python
new_dataframe['tags'] = new_dataframe['tags'].apply(lemmatize_text)
```

```python
nltk.download('stopwords')
```

```python
# Define function to preprocess text (remove stopwords)
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    words = text.split()
    clean_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(clean_words)
```

```python
# Apply preprocessing
new_dataframe['tags'] = new_dataframe['tags'].apply(preprocess_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_dataframe['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

        
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_dataframe[new_dataframe['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new_dataframe[new_dataframe['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_dataframe.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```
